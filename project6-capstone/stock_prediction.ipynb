### US Stock Fundamentals From Morningstar  ###

```python
api_fundamentals = 'http://financials.morningstar.com/ajax/exportKR2CSV.html?t=QCOM'

import pandas as pd
import urllib.request as r
import sys

if sys.version_info[0] < 3: 
    from StringIO import StringIO
else:
    from io import StringIO

request = r.Request(api_fundamentals)
response = r.urlopen(request)
txt = StringIO(response.read().decode('utf-8'))
#print(response.read().decode('utf-8'))

data = pd.read_csv(txt, sep=",", skiprows=2)
data[:5]
```
